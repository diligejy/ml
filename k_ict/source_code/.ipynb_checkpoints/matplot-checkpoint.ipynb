```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x), '-', linewidth=1, label='sin(x)')
plt.plot(x, np.cos(x), '--', linewidth=1, label='cos(x)')
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.title('Sine & Cosine Curve')
plt.xlabel('x')
plt.legend()
```
